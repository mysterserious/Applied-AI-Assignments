```python
def draw_line(coef,intercept, mi, ma):
    # for the separating hyper plane ax+by+c=0, the weights are [a, b] and the intercept is c
    # to draw the hyper plane we are creating two points
    # 1. ((b*min-c)/a, min) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the minimum value of y
    # 2. ((b*max-c)/a, max) i.e ax+by+c=0 ==> ax = (-by-c) ==> x = (-by-c)/a here in place of y we are keeping the maximum value of y
    points=np.array([[((-coef[1]*mi - intercept)/coef[0]), mi],[((-coef[1]*ma - intercept)/coef[0]), ma]])
    plt.plot(points[:,0], points[:,1])
```

# **TASK 1** Applying SVM

```python
ratios = [(100,2), (100, 20), (100, 40), (100, 80)]
c=[0.001, 1, 100]
plt.figure(figsize=(20,10))
count=0
data_x=[]
data_y=[]
for j,i in enumerate(ratios):
        X_p=np.random.normal(0,0.05,size=(i[0],2))
        X_n=np.random.normal(0.13,0.02,size=(i[1],2))
        y_p=np.array([1]*i[0]).reshape(-1,1)
        y_n=np.array([0]*i[1]).reshape(-1,1)
        X=np.vstack((X_p,X_n))
        y=np.vstack((y_p,y_n))
        data_x.append((X_p,X_n))
        data_y.append((y_p,y_n))
        for a in range(len(c)):
            count=count+1
            plt.subplot(4, 3, count)
            clf1=SVC(kernel='linear',C=c[a],random_state=0)
            clf1.fit(X,y)
            coef=clf1.coef_
            intercept=clf1.intercept_
            draw_line(coef[0], intercept, mi=min(X[:,1]), ma=max(X[:,1]))
            plt.scatter(X_p[:,0],X_p[:,1])
            plt.scatter(X_n[:,0],X_n[:,1],color='red')
            plt.suptitle('Hyperplane plots with different values of hyperparameter and imbalanced datasets using Linear SVM')
plt.show()
```

# **Observation:**
1. With increase in C, loss functions gets dominated and the model tends to overfitting.

2. From the above plots we can say that, with increase in C, model is moving towards well fit from underfit and if we further increase the parameter it will leads to overfitting of the model.

3.After looking at the plots I can tell both imbalanced data and the hyperparameter C can affect the orientation of the best hyperplane. When the data is highly imbalanced then we would have to increase our hyperparameter C to get the optimal hyperplane. 

4. At C=0.001, in all cases model is underfit.

5. At C=100, in all cases model is optimal.
