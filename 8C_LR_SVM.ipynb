```python
# you should not import any other packages
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
from sklearn.linear_model import SGDRegressor
```

```python
import numpy as np
import scipy as sp
import scipy.optimize

def angles_in_ellipse(num,a,b):
    assert(num > 0)
    assert(a < b)
    angles = 2 * np.pi * np.arange(num) / num
    if a != b:
        e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
        tot_size = sp.special.ellipeinc(2.0 * np.pi, e)
        arc_size = tot_size / num
        arcs = np.arange(num) * arc_size
        res = sp.optimize.root(
            lambda x: (sp.special.ellipeinc(x, e) - arcs), angles)
        angles = res.x 
    return angles
```

```python
a = 2
b = 9
n = 50

phi = angles_in_ellipse(n, a, b)
e = (1.0 - a ** 2.0 / b ** 2.0) ** 0.5
arcs = sp.special.ellipeinc(phi, e)

fig = plt.figure()
ax = fig.gca()
ax.axes.set_aspect('equal')
ax.scatter(b * np.sin(phi), a * np.cos(phi))
plt.show()
```

```python
X= b * np.sin(phi)
Y= a * np.cos(phi)
```

```python
outliers=[(0,2),(21, 13), (-23, -15), (22,14), (23, 14)]
alpha=[0.0001, 1, 100]
count=0
plt.figure(figsize=(20,10))
for i in alpha:
  X= b * np.sin(phi)
  Y= a * np.cos(phi)
  for j in outliers:
    X=np.append(X,j[0])
    Y=np.append(Y,j[1])
    count+=1
    clf=SGDRegressor(loss='squared_error',penalty='l2',alpha=i, eta0=0.001, learning_rate='constant',random_state=0)
    clf.fit(X.reshape(-1,1),Y)
    y_pred=clf.predict(X.reshape(-1,1))
    plt.subplot(3,5,count)
    plt.suptitle('Hyperplane plots with different values of hyperparameter and different number of outliers')
    plt.scatter(X,Y)
    plt.plot(X,y_pred)
plt.show()    
```

1. In the above task we can see that with the increase in number of outliers and keeping the hyperparameter same, the model tends to overfitting.
