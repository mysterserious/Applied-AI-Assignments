```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore") 
import matplotlib.pyplot as plt
```

# ***TASK 1***

```python
data=pd.read_csv('5_a.csv')
print(data.head(10))
```

```python
probability=data['proba']
print(probability)
```

```python
y_pred=[]
count=0
for i in range(len(probability)):
    if probability[i]<0.5:
        count+=1
        y_pred.append(0)
    else:
        y_pred.append(1)
print(y_pred)
print(count)          
```

```python
del data['proba']
print(data.head(10))
```

```python
data['y_pred']=y_pred
print(data.head(10))
```

```python
tp=0
fp=0
tn=0
fn=0
for i in range(len(data['y'])):
    if data['y'][i]==1 and data['y_pred'][i]==1:
        tp+=1
    elif data['y'][i]==0 and data['y_pred'][i]==0:
        tn+=1
    elif data['y'][i]==1 and data['y_pred'][i]==0:
        fn+=1 
    elif data['y'][i]==0 and data['y_pred'][i]==1:
        fp+=1
print(tp)
print(tn)
print(fp)
print(fn)
```

```python
values=[[tn,fn],[fp,tp]]
confusion_mat=np.array([np.array(i) for i in values])
print('Confusion Matrix of dataset a :')
print(confusion_mat)
```

```python
pr_a=float(tp/(tp+fp))
print('Precision :',pr_a)
re_a=float(tp/(fn+tp))
print('Recall :',re_a)
f1_a=float((2*pr_a*re_a)/(pr_a+re_a))
print('F1 score of dataset a is',f1_a)
```

```python
data['proba']=round(probability,2)
unique_prob=list(set(data['proba']))
unique_prob=sorted(unique_prob,reverse=True)
print(unique_prob)
```

```python
tprs=[]
fprs=[]
for i in unique_prob:
    y_hat=[]
    t=i
    for j in range(len(data['y'])):
        if data['proba'][j]<t:
            y_hat.append(0)
        else:
            y_hat.append(1) 
    data['y_hat']=y_hat
    tp1=len(data[(data['y']==1) & (data['y_hat']==1)])
    fp1=len(data[(data['y']==0) & (data['y_hat']==1)])
    fn1=len(data[(data['y']==1) & (data['y_hat']==0)])
    tn1=len(data[(data['y']==0) & (data['y_hat']==0)])
    tpr=tp1/(tp1+fn1)
    fpr=fp1/(tn1+fp1)
    tprs.append(tpr)
    fprs.append(fpr)
auc=np.trapz(tprs,fprs)
print('AUC of dataset a is :',auc)
    
```

```python
plt.plot(fprs,tprs)
plt.xlabel('FPRs')
plt.ylabel('TPRs')
plt.title('ROC curve of dataset a')
plt.show()
```

```python
acc_score1=(tp+tn)/len(data['y'])
print('Accuracy score of dataset a is',round(acc_score1,ndigits=2))
```

```python
df_d=pd.read_csv('5_d.csv')
df_d.tail(10)
```

```python
sum=0
for i in range(len(df_d['y'])):
    er=float(df_d['y'][i]-df_d['pred'][i])
    sum+=float(er**2)
mean_er=sum/len(df_d)
mean_er=round(mean_er,ndigits=2)    
print('Mean Square Error of dataset d is',mean_er)
```

```python
y_sum=0
er_sum=0
for i in range(len(df_d['y'])):
    er=float(df_d['y'][i]-df_d['pred'][i])
    er_sum+=abs(er)
    y_sum+=df_d['y'][i]
mape=er_sum/y_sum
print('MAPE of dataset d is',mape)
```

```python
y_mean=df_d['y'].sum()/len(df_d['y'])
ss=0
y_mean=round(y_mean,ndigits=2)
for i in range(len(df_d['y'])):
    diff=df_d['y'][i]-y_mean
    ss+=diff**2
ss=round(ss,ndigits=2)
r_sq=1-(sum/ss)
r_sq=round(r_sq,ndigits=2)
print('r^2 error of dataset d is',r_sq)    
```
